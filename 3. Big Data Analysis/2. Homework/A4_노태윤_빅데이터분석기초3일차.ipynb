```python
import pandas as pd
import numpy as np
```

```python
#데이터 구성하기 
df1 = pd.read_csv('/home/piai/data Handling/data/FITNESS.csv',encoding ='euc-kr')
```

```python
df1.head()
```

```python
#groupby,agg를 이용한 평균값 계산
df_new = df1.groupby(['GENDER','AGEGROUP'])['WEIGHT'].agg(['mean'])
```

```python
pd.DataFrame.pivot_table(df_new, index = 'GENDER', columns = 'AGEGROUP', values = None)
```

```python
pd.DataFrame.pivot_table(df_new, index = 'AGEGROUP', columns = 'GENDER', values = None)
```
