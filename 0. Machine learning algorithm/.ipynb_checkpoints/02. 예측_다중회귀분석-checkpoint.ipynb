# 다중회귀분석 실습

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 회귀분석
from statsmodels.formula.api import ols
# qqplot, 회귀분석의 상수항 추가
from statsmodels.api import qqplot, add_constant
# 선형 모델 formula(y ~ X1 + X2 + ...)
import statsmodels.formula.api as smf
# 다중공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor
# 선형회귀모형
from sklearn.linear_model import LinearRegression
# 변수 선택법(후진제거법)
from sklearn.feature_selection import RFE
# Scale 변환->표준화 회귀계수 산출
from sklearn.preprocessing import StandardScaler
# 평가함수
from statsmodels.tools.eval_measures import rmse
# jupyter notebook 사용 시 그래프 자동 출력 옵션
%matplotlib inline

import matplotlib
# 한글 폰트 적용
matplotlib.rc('font', family = 'Nanumgothic')

# 음수 표시
matplotlib.rc('axes', unicode_minus=False)
```

```python
df_raw = pd.read_csv('./Data/체질검사.csv', engine = 'python')
df_raw.head()
```

```python
df_raw['WEIGHT'] = df_raw['WEIGHT'] * 0.45
df_raw['HEIGHT'] = df_raw['HEIGHT'] * 2.54
df_raw['BMI'] = df_raw['WEIGHT'] / ((df_raw['HEIGHT']*0.01) **2)
```

```python
df_raw.head()
```

FAT을 예측하기 위해 BMI라는 파생변수를 생성했으며, 보기 쉽게 WEIGHT를 kg단위로, HEIGHT를 cm단위로 변경하였다.

```python
sns.pairplot(df_raw)
```

전체 변수에 대한 관계를 scattor plot으로 표현했는데, 선형관계를 보이는 변수들이 다수 확인된다. 이는 변수들 간의 다중공선성도 어느 정도 있을 것으로 예상된다. 더 자세한 분석을 위해 상관관계를 수치로 표현해보자.

```python
df_raw.corr().round(3)
```

WEIGHT, ABDOMEN, HIP, THIGH 등의 설명변수들이 FAT이라는 목표변수와 상관관계가 있는 것으로 확인된다.   
<br>
다만 ABDOMEN과 CHEST 간의 상관관계가 0.916으로 매우 높게 나오며, 그 이외에도 설명변수들 간의 선형관계도 확인되므로 다중공선성에 대한 처리가 필요하다고 판단된다.
<br><br>

```python
reg_model = smf.ols(formula = 'FAT~AGE+WEIGHT+HEIGHT+NECK+CHEST+ABDOMEN+HIP +\
+THIGH+KNEE+ANKLE+BICEPS+FOREARM+WRIST+BMI', data = df_raw)

reg_result = reg_model.fit()
reg_result.summary()
```

전체 변수에 대한 회귀모델을 생성하였다.  
우선 해당 모델에 대한 유의성 검정인 F검정의 p-value 값이 거의 0에 가까운 것으로 보아 해당 모델은 회귀모델로서 유의하다고 볼 수 있다.  
수정결정계수( Adj. R-squared )는 0.734로 이 모델을 통하여 설명변수를 73% 정도 설명할 수 있으므로 모델 성능이 꽤 좋다고 판단된다.  
Prob(Omnibus)값이	0.118로 유의수준을 0.05라고 보았을 때 잔차의 정규성 역시 만족되는 것으로 보인다. Durbin-Watson 값이 2.113이므로 잔차의 독립성도 만족한다.

전체 변수에 대한 회귀모델도 훌륭하지만 앞서 말했듯이 설명변수들 간의 다중공선성이 확인되므로 이를 확인한 후 제거하는 과정을 진행한다.

```python
df_raw_x = df_raw.drop('FAT', axis = 1)
df_raw_x_const = add_constant(df_raw_x)

df_vif = pd.DataFrame()

df_vif['variable'] = df_raw_x_const.columns # column 값들을 넣음
df_vif['VIF'] = [variance_inflation_factor(df_raw_x_const.values, i) \
                for i in range(df_raw_x_const.shape[1])]
df_vif.sort_values('VIF', inplace = True)
df_vif.round(3)
```

VIF값이 10이상이면 보통 다중공선성이 있다고 판단하는데 WEIGHT의 경우 52.665로 매우 높게 나왔기 때문에 제거가 필요하다.

```python
df_raw_x = df_raw.drop(['FAT', 'WEIGHT'], axis = 1)
df_raw_x_const = add_constant(df_raw_x)
df_vif = pd.DataFrame()
df_vif['variable'] = df_raw_x_const.columns
df_vif['VIF'] = [variance_inflation_factor(df_raw_x_const.values, i) for i in range(df_raw_x_const.shape[1])]
df_vif.sort_values('VIF', inplace = True)
df_vif.round(3)
```

ABDOMEN과 HIP 모두 여전히 VIF값이 10을 넘지만, 앞서서 FAT과의 상관관계를 확인했을 때, 이 두 값 모두 목표변수와의 상관성이 높았기 때문에 중요변수일 확률이 높아서 제거하지 않는다.

```python
reg_model = smf.ols(formula = 'FAT~AGE+HEIGHT+NECK+CHEST+ABDOMEN+HIP +\
+THIGH+KNEE+ANKLE+BICEPS+FOREARM+WRIST+BMI', data = df_raw)
reg_result = reg_model.fit()
reg_result.summary()
```

설명변수 중 WEIGHT 값을 제거한 후에도 수정 결정계수인 Adj. R-squared 값이 0.733으로 앞선 전체 설명변수에 대한 값인 0.734과 거의 차이가 없는 것을 볼 수 있다. 즉, WEIGHT는 목표변수에 큰 영향을 주지 않는 것으로 판단할 수 있다.

후진제거법으로 변수를 선택하기 이전에 p-value가 0.05이하인 설명변수들만 선택한 회귀모델을 한 번 실행해보자.  
t검정에 대해 p-value가 0.05보다 작다는 것은 해당 회귀계수가 유의하다는 의미이기 때문에 p-value값으로 판단한 모델과 후진제거법으로 생성한 모델 간의 차이를 비교하기 위해 해당 과정을 진행해본다. (그냥 궁금해서 진행해봤습니다!)


```python
# p-value가 0.05이하인 설명변수들 : AGE NECK ABDOMEN  HIP FOREARM WRIST
    
reg_model = smf.ols(formula = 'FAT~AGE + NECK + ABDOMEN + HIP + FOREARM + WRIST', data = df_raw)
reg_result = reg_model.fit()
reg_result.summary()
```

해당 모델은 모델의 유의성, 잔차의 정규성, 독립성을 모두 만족하며, 수정 결정계수 값이 0.730으로 이전 모델들보다 아주 조금 낮아졌지만 거의 비슷한 것을 볼 수 있다.

후진제거법을 이용한 모델

```python
df_raw_y = df_raw['FAT']
model = LinearRegression()
rfe = RFE(estimator = model, n_features_to_select = 5).fit(df_raw_x, df_raw_y)
selected_cols = df_raw_x.columns[rfe.support_]
removed_cols = df_raw_x.columns[~rfe.support_]

print('Selected Variables: {}'.format(selected_cols))
print('Removed Variables: {}'.format(removed_cols))
```

```python
# 'NECK', 'ABDOMEN', 'HIP', 'FOREARM', 'WRIST'
rfe_reg_model = smf.ols(formula = 'FAT~NECK + ABDOMEN + HIP + FOREARM + WRIST', data = df_raw)
rfe_reg_result = rfe_reg_model.fit()
rfe_reg_result.summary()
```

해당 모델에 대한 수정 결정계수는 0.726로 작은 차이이긴 해도 지금까지의 모델 중에 성능이 가장 낮은 것을 볼 수 있다.   
즉, 후진제거법으로 구한 모델보다 p-value로 판단한 모델의 성능이 더 좋은 것을 볼 수 있다.

앞선 모델들을 살펴본 결과 최종적으로 p-value로 판단한 모델을 채택했다.  
마지막으로 해당 모델의 설명 변수 중요도를 살펴본다. 

```python
scaler = StandardScaler()
df_raw_p = df_raw_x.drop(['HEIGHT', 'CHEST', 'THIGH', 'KNEE', 'ANKLE', 'BICEPS', 'BMI'], axis = 1)
cols = df_raw_p.columns
np_scaled = scaler.fit_transform(df_raw_p)
df_scaled = pd.DataFrame(np_scaled, columns = cols)

df_scaled['FAT'] = df_raw['FAT']
reg_model_scaled = smf.ols(formula = 'FAT~AGE + NECK + ABDOMEN + HIP + FOREARM + WRIST', data = df_scaled)
reg_result_scaled = reg_model_scaled.fit()
reg_result_scaled.summary()
```

```python
df_reg_coef = pd.DataFrame({'Coef': reg_result_scaled.params.values[1:]}, index = cols)
df_reg_coef.plot.barh(y = 'Coef', legend = False)
```

> 결론
목표변수 FAT와 설명변수들 간의 회귀분석을 진행해본 결과, 각각의 회귀계수에 대한 p-value 값으로 판단한 모델을 선정했습니다. 해당 모델을 사용했을 때 FAT에 대한 설명변수는 WRIST, FOREARM, HIP, ABDOMEN, NECK. AGE 총 6개이며 이 중에서 ABDOMEN 값이 가장 변수 중요도가 높은 것을 볼 수 있었습니다.  
<br>
또한, 파생변수로 생성했던 BMI는 생각보다 FAT과의 상관성이 높지 않았습니다.   
<br>
최종 모델의 회귀식 : y = -1.81 + 0.026 AGE -0.576 NECK + 0.894 ABDOMEN -0.213 HIP + 0.5024 - 1.961 WRIST

<br>



> 효율성 관점에서 생각했을 때는 '후진제거법을 이용한 모델이 더 좋지 않을까'하는 생각도 듭니다. 5개 변수를 사용해서 adj R-squared값이 0.726인 것과 6개 변수를 사용해서 adj R-squared값이 0.730이기 때문에 R-squared 값이 차이가 크지 않은만큼 후진제거법을 이용한 모델을 사용해도 좋을 것 같습니다.
